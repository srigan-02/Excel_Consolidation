```python
## What it does is
- Loads all sheets.
- Identifies headers that exist in all sheets.
- Filters each sheet to retain only those headers.
- Concatenates all filtered data into one DataFrame.
- Saves the final result to consolidated_common_headers.xlsx.
```

```python
import pandas as pd

# Corrected file path
file_path = r"path"   # Replace with your actual file path

# Load the Excel file
xls = pd.ExcelFile(file_path)

# Read all sheets into a list of DataFrames
dataframes = [xls.parse(sheet) for sheet in xls.sheet_names]

# Find common columns across all sheets
common_columns = set(dataframes[0].columns)
for df in dataframes[1:]:
    common_columns.intersection_update(df.columns)

# Convert set to list and ensure consistent column order
common_columns = list(common_columns)

# Filter each DataFrame to keep only the common columns
filtered_dfs = [df[common_columns].copy() for df in dataframes]

# Concatenate all filtered DataFrames
consolidated_df = pd.concat(filtered_dfs, ignore_index=True)

# Save to a new Excel file
consolidated_df.to_excel("consolidated_common_headers.xlsx", index=False)

print("Consolidation complete. Output saved to 'consolidated_common_headers.xlsx'.")
```
